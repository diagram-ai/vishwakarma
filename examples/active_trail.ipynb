```python
from pgmpy.models import BayesianModel
from pgmpy.models import NaiveBayes
from pgmpy.models import MarkovModel
from pgmpy.models import FactorGraph
from pgmpy.models import DynamicBayesianNetwork
import pgmpy
import numpy as np
import vishwakarma
from IPython.display import Image
```

```python
model = pgmpy.models.BayesianModel([('Fraud', 'Gas'),
                       ('Fraud', 'Jewellery'),
                       ('Age', 'Jewellery'),
                       ('Sex', 'Jewellery')])
```

```python
model_viz = vishwakarma.pgmplot(model)
display(model_viz) 
```

```python
imgrend = vishwakarma.activetrail(model,"Jewellery","Age")
```

```python
display(imgrend) 
```
