```python
import pgmpy
import numpy as np
import vishwakarma
from IPython.display import Image
```

```python
# Bayesian Model
model = pgmpy.models.BayesianModel([('Fraud', 'Gas'),
                       ('Fraud', 'Jewellery'),
                       ('Age', 'Jewellery'),
                       ('Sex', 'Jewellery')])

model_viz = vishwakarma.pgmplot(model)
display(model_viz)
```

```python
# Naive Bayes model
model = pgmpy.models.NaiveBayes([('a', 'b'),
                                 ('a', 'c'),
                                 ('a', 'd')])

model_viz = vishwakarma.pgmplot(model)
display(model_viz)
```

```python
# Markov Model
model = pgmpy.models.MarkovModel([('A', 'B'),
                                  ('B', 'C'),
                                  ('C','D'),
                                  ('D','A')])

model_viz = vishwakarma.pgmplot(model)
display(model_viz)
```

```python
# Factor Graph
model = pgmpy.models.FactorGraph()
model.add_nodes_from(['a', 'b', 'c'])
phi1 = pgmpy.factors.discrete.DiscreteFactor(['a', 'b'], [2, 2], np.random.rand(4))
phi2 = pgmpy.factors.discrete.DiscreteFactor(['b', 'c'], [2, 2], np.random.rand(4))
model.add_nodes_from([phi2, phi1])
model.add_factors(phi1, phi2)
model.add_edges_from([('a', phi1), ('b', phi1),('b', phi2), ('c', phi2)])

model_viz = vishwakarma.pgmplot(model)
display(model_viz)
```

```python
# Dynamic Bayesian
model = pgmpy.models.DynamicBayesianNetwork()
model.add_edges_from([(('D', 0), ('G', 0)),
                      (('I', 0), ('G', 0)),
                      (('G', 0), ('L', 0)),
                      (('D', 0), ('D', 1)),
                      (('I', 0), ('I', 1)),
                      (('G', 0), ('G', 1)),
                      (('G', 0), ('L', 1)),
                      (('L', 0), ('L', 1)),
                      (('G', 1), ('L', 2))])

model_viz = vishwakarma.adapter.pgmpy.DynamicBayesianNetwork(model)
imgrend = vishwakarma.renderer.latex.ImageRenderer(model_viz.model_json)
display(imgrend.image)

model.add_node('S')
model.add_edges_from([(('S',0),('S',1)),
                      (('I',0),('S',0)),
                      (('I',1),('S',1))])

model_viz = vishwakarma.pgmplot(model)
display(model_viz)
```
